### Importing the Data

```python
import pandas as pd

# Read the CSV file, treating 'Date' column as a string initially
kseData = pd.read_csv("./1/Stock Exchange KSE 100(Pakistan).csv")

# Convert the 'Date' column to datetime explicitly with the correct format
kseData['Date'] = pd.to_datetime(kseData['Date'], format='%d-%b-%y', errors='coerce')

# Set the 'Date' column as the index
kseData.set_index('Date', inplace=True)

# Remove commas from numeric columns and convert them to appropriate types
numeric_columns = ['Open', 'High', 'Low', 'Close', 'Change', 'Volume']
kseData[numeric_columns] = kseData[numeric_columns].replace({',': ''}, regex=True)
kseData[numeric_columns] = kseData[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
kse_close = kseData['Close']
kse_close.reset_index(drop= True, inplace= True)
kse_close.shape
```

```python
kse_close = pd.DataFrame(kse_close)
kse_close.head()
```

#### Pre-pre-processing the Data

```python
# Split data into train and test, ensuring no overlap
train_size = int(len(kse_close) * 0.8)  # 80% train, 20% test
train = kse_close.iloc[:train_size]
test = kse_close.iloc[train_size:]
```

```python
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()
scalar.fit(train)

scaled_train = scalar.transform(train)
scaled_test = scalar.transform(test)
```

### RNN Model

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

print(test.shape)

length = 16
batch_size = 32

generator = TimeseriesGenerator(scaled_train, scaled_train, length = length, batch_size = batch_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, Dropout

n_features = 1

model = Sequential()
model.add(SimpleRNN(30, input_shape=(length, n_features)))
model.add(Dense(1))

model.compile(optimizer= 'adam', loss= "mse")

model.summary()
```

```python
model.fit(generator, epochs= 20, shuffle=False)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot(figsize=(6,3))
```

### Evaluting on test set - RNN

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Generate predictions on the test set using TimeseriesGenerator
test_generator = TimeseriesGenerator(scaled_test, scaled_test, length=length, batch_size=batch_size)

# Predict the next values using the trained model
predictions = model.predict(test_generator)

# Inverse transform the predictions to get the actual values in the original scale
predicted_values_actual_scale = scalar.inverse_transform(predictions)

# Extract the actual values from the test set (the next 7 days)
actual_values = scaled_test[length:length + len(predictions)]

# Inverse transform the actual values to get them in the original scale
actual_values_actual_scale = scalar.inverse_transform(actual_values)

# # Print the predicted values vs the actual values
# print("Predicted values:", predicted_values_actual_scale.flatten())
# print("Actual values:", actual_values_actual_scale.flatten())

# Calculate RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(actual_values_actual_scale, predicted_values_actual_scale))
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Calculate MAE (Mean Absolute Error)
mae = mean_absolute_error(actual_values_actual_scale, predicted_values_actual_scale)
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
import matplotlib.pyplot as plt

plt.plot(actual_values_actual_scale, label='Actual values')
plt.plot(predicted_values_actual_scale, label='Predicted values')
plt.legend()
plt.title("Predicted vs Actual values")
plt.show()
```

### LSTM Model

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor = "val_loss", patience=4)
```

```python
length = 15

generator = TimeseriesGenerator(scaled_train, scaled_train, length = length, batch_size= 32)

validation_generator = TimeseriesGenerator(scaled_test, scaled_test, length = length, batch_size= 32)
```

```python
model = Sequential()
model.add(LSTM(30, input_shape=(length, n_features)))
model.add(Dropout(rate=0.2))
model.add(Dense(1))

model.compile(optimizer= 'adam', loss= "mse")

model.summary()
```

```python
model.fit(generator, epochs = 20, validation_data= validation_generator, callbacks=[early_stop], shuffle=False)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
# Generate predictions on the test set using TimeseriesGenerator for LSTM model
lstm_predictions = model.predict(validation_generator)

# Inverse transform the predictions to get them in the original scale
lstm_predictions_actual_scale = scalar.inverse_transform(lstm_predictions)

# Extract actual values from the test set
actual_values = scaled_test[length:length + len(lstm_predictions)]

# Inverse transform actual values to get them in the original scale
actual_values_actual_scale = scalar.inverse_transform(actual_values)

# Compare predictions with actual values
# # Plot actual values vs. predictions from both RNN and LSTM
plt.plot(actual_values_actual_scale, label='Actual values')
plt.plot(predicted_values_actual_scale, label='RNN Predictions')
plt.plot(lstm_predictions_actual_scale, label='LSTM Predictions')
plt.legend()
plt.title("Actual vs RNN vs LSTM Predictions")
plt.show()
# Calculate RMSE for LSTM
lstm_rmse = np.sqrt(mean_squared_error(actual_values_actual_scale, lstm_predictions_actual_scale))
print(f"LSTM Root Mean Squared Error (RMSE): {lstm_rmse}")

# Calculate MAE for LSTM
lstm_mae = mean_absolute_error(actual_values_actual_scale, lstm_predictions_actual_scale)
print(f"LSTM Mean Absolute Error (MAE): {lstm_mae}")

```

```python
# Plot actual values vs. predictions from both RNN and LSTM
plt.plot(actual_values_actual_scale, label='Actual values')
plt.plot(predicted_values_actual_scale, label='RNN Predictions')
plt.plot(lstm_predictions_actual_scale, label='LSTM Predictions')
plt.legend()
plt.title("Actual vs RNN vs LSTM Predictions")
plt.show()
```
